```python
import pandas as pd
df = pd.read_csv("titanic.csv")
df.head()
```

```python
yDf = df["Survived"]
xDf = df.drop("Survived", axis=1)
```

```python
columnsDropped = [
    "Name",
    "Ticket",
    "Embarked",
    "PassengerId",
]
xDf = xDf.drop(columns=columnsDropped)
xDf.head()
```

```python
xDf["hasCabin"] = xDf["Cabin"].notna().astype(int)
xDf = xDf.drop(columns=["Cabin"])
```

```python
xDf["Sex"] = xDf["Sex"].map({"male": 0, "female": 1}).astype(int)
xDf.head()
```

```python
xDf.info()
```

```python
xDf["Age"] = xDf["Age"].fillna(xDf["Age"].mean())
xDf.info()
```

```python
from sklearn.preprocessing import StandardScaler

scaler_X = StandardScaler()
xDf = scaler_X.fit_transform(xDf)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(xDf, yDf, test_size=0.2, random_state=42)

from sklearn.linear_model import LogisticRegression
model = LogisticRegression()

model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)

from sklearn.metrics import accuracy_score, confusion_matrix , classification_report
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn_model = KNeighborsClassifier(n_neighbors= 5)
knn_model.fit(X_train,y_train)
```

```python
y_pred_knn = knn_model.predict(X_test)
```

```python
accuracy_score(y_test,y_pred_knn)
```

```python
confusion_matrix(y_test,y_pred_knn)
```

```python
print(classification_report(y_test,y_pred_knn))
```

```python
from sklearn.naive_bayes import GaussianNB
model_NB = GaussianNB()

model_NB.fit(X_train,y_train)

y_pred_NB = model_NB.predict(X_test)
```

```python
accuracy_score(y_test,y_pred_NB)
```

```python
confusion_matrix(y_test,y_pred_NB)
```

```python
print(classification_report(y_test,y_pred_NB))
```

```python
from sklearn.tree import DecisionTreeClassifier
model_DT = DecisionTreeClassifier(random_state=42)

model_DT.fit(X_train,y_train)
```

```python
y_pred_DT = model_DT.predict(X_test)
accuracy_score(y_test,y_pred_DT)
```

```python
from sklearn.svm import SVC
model_svm = SVC(kernel = 'linear')
model_svm.fit(X_train, y_train)
```

```python
y_pred_svc = model_svm.predict(X_test)
accuracy_score(y_test,y_pred_svc)
```

```python
confusion_matrix(y_test,y_pred_svc)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_score

scaler = StandardScaler()
X_scaled = scaler.fit_transform(xDf)

scores = cross_val_score(
    model_svm,
    X_scaled,
    yDf,
    cv=5,
    scoring='accuracy'
)

print(scores.mean())

# What it does in one line:
# Trains and tests the SVM 5 times on different folds and returns 5 accuracy scores.
```
